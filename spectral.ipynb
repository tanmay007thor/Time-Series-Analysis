```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

t = np.linspace(0, 10, num=1000)
signal = np.sin(2 * np.pi * 1 * t) + 0.5 * np.sin(2 * np.pi * 5 * t) 
noise = np.random.normal(0, 0.1, size=1000)  
signal = signal + noise

fft_result = np.fft.fft(signal)
fft_frequency = np.fft.fftfreq(len(signal), 1 / 100)

cutoff_frequency = 6.0 
filtered_fft_result = fft_result.copy()
filtered_fft_result[np.abs(fft_frequency) > cutoff_frequency] = 0

plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.xlabel('Time')
plt.ylabel('Signal')
plt.title('Signal with Noise')

print(fft_frequency)

plt.subplot(2, 1, 2)
plt.plot(fft_frequency, np.abs(fft_result))
plt.xlabel('Frequency ')
plt.ylabel('Magnitude')
plt.title('FFT Result')
plt.xlim(-10, 10) 
plt.tight_layout() 
plt.show()
```
